```python
# Import libraries
import glob
import pandas as pd
```

```python
path =r'/Users/cephra.stuart/Desktop/CitiBikeData' # use your path
allFiles = glob.glob(path + "/*.csv")

list_ = []

for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)

frame = pd.concat(list_, axis = 0, ignore_index = True)
```

```python
df = frame
df.head()
```
